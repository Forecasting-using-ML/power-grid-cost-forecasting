```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
results = {
    'ANN': 'ann',
    'Bayesian Ridge': 'br',
    'Decision Tree': 'dt',
    'Dummy': 'dummy',
    'Gradient Boosting': 'gb',
    'GRU': 'gru',
#     '$k$NN': 'knn',
    'LSTM': 'lstm',
    'Passive Aggressive Regressor': 'pagg',
    'Random Forest': 'rf',
    'Simple RNN': 'simple-rnn'
}
```

```python
dfs = []

for name, prefix in results.items():
    filename = 'results/{}-mse.csv'.format(prefix)
    df = pd.read_csv(filename, index_col=[0], names=[name]).T
    dfs.append(df)
    
df = pd.concat(dfs)
df.name = 'MSE'
```

```python
df.sort_values('Test')
```

## Train

```python
ax = df['Train'].sort_values().plot(kind='bar')
_ = ax.set_ylabel('MSE')
```

```python
ax = df['Train'].sort_values()[:5].plot(kind='bar')
_ = ax.set_ylabel('MSE')
```

## Test

```python
ax = df['Test'].sort_values().plot(kind='bar')
_ = ax.set_ylabel('MSE')
```

```python
ax = df['Test'].sort_values()[:5].plot(kind='bar')
_ = ax.set_ylabel('MSE')
```

## Holistic

```python
ax = df.sort_values(['Test']).plot(kind='bar')
_ = ax.set_ylabel('MSE')
```

```python
ax = df.sort_values(['Test'])[:5].plot(kind='bar')
_ = ax.set_ylabel('MSE')
```
