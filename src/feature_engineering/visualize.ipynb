# Train

```python
train = pd.read_csv('train.gz')
```

```python
_ = heatmap(train)
```

### $X$

```python
X_train = train[train.columns[:-1]]
```

```python
X_train.head()
```

```python
X_train.tail()
```

```python
axs = X_train[X_train.columns[5:]].plot(figsize=(16, 8), subplots=True)
for ax in axs:
    ax.set_ylabel('MWHr')
    ax.set_xlabel('Time Step')
```

### $y$

```python
y_train = train[train.columns[-1]]
```

```python
y_train.head()
```

```python
y_train.head()
```

```python
ax = y_train.plot(figsize=(16, 4))
ax.set_ylabel('MWHr/$')
_ = ax.set_xlabel('Time Step')
```
